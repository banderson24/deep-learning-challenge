## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'STATUS'])
application_df.head()
```

```python
application_df.dtypes
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
print(unique_values)
```

```python
# Look at APPLICATION_TYPE value counts for binning
apptype_column = application_df['APPLICATION_TYPE'].value_counts()
print(apptype_column)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
other = 400
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < other].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_column = application_df['CLASSIFICATION'].value_counts()
print(classification_column)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
print(classification_column[classification_column>1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
other = 2
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() < other].index.tolist()


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Look at NAME counts for binning
name_column = application_df['NAME'].value_counts()
print(name_column)
```

```python
# Choose a cutoff value and create a list of names to be replaced
# use the variable name `names_to_replace`
other = 10
names_to_replace = application_df['NAME'].value_counts()[application_df['NAME'].value_counts() < other].index.tolist()


# Replace in dataframe
for cls in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
unique_values_new = application_df.nunique()
print(unique_values_new)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
number_input_features = len( X_train_scaled[0])
print(number_input_features)
```

```python
# Determine how many input features there are
num_input_features = application_df.shape[1] - 1
print("Number of input features:", num_input_features)
```

## First Optimization Attempt
### For this 1st optimization attempt I reduced the cutoff point for the APPLICATION_TYPE column from 528 to 400 creating fewer bins to look at. I also reduced the cutoff point for the CLASSIFICATION column from 1883 to 10 to increase the number of data points we looked at. I also tried several different feature combinations and eventually landed on removing EIN and STATUS. After adding NAME back I saw a large increase in results so I added that back and binned the groups to only look at NAMEs with more than 10 occurances as my colab was crashing if I tried to do all the names at once. I also used the same number of neurons as input features.

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = application_df.shape[1] - 1

neurons = input_features


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=neurons, activation="relu", input_dim=input_features))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Second Optimization Attempt
### For the 2nd optimization attempt I tried doubling the number of neurons I was using in the model to see if that would have an impact on the accuracy score. I also added a hidden layer to see how that might impact the code.

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = application_df.shape[1] - 1
# In class we learned we should use 2-3 times the # of input features for neurons for an initial model

# Same neuron number as that seemed to ever so slightly improve it
neurons1 = input_features*2

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(tf.keras.layers.Dense(units=neurons1, activation="relu", input_dim=input_features))

# Second Hidden Layer
nn2.add(tf.keras.layers.Dense(units=neurons1, activation="relu"))


# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn2.summary()
```

```python
# Compile the model
nn2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn2.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Third Optimization Attempt
### For the third Optimization Attempt I created a method that allowed kerastuner to decide the optimal activation function, number of neurons, and hidden layers.

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):

    input_features = application_df.shape[1] - 1
    nn3 = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn3.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=400,
        step=5), activation=activation, input_dim=input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn3.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=400,
            step=5),
            activation=activation))

    nn3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn3.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn3
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Search for the best hyperparameters
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
# Evaluate best model against full test data
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
